```python
# df['Last_Name'] = df['Last_Name'].str.strip('/')
# df['Last_Name'] = df['Last_Name'].str.strip('...')
# df['Last_Name'] = df['Last_Name'].str.strip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('/','')
df['Phone_Number'] = df['Phone_Number'].str.replace('-','')
df['Phone_Number'] = df['Phone_Number'].str.replace('|','')
```

```python
df['Phone_Number'].replace('Na',np.nan, inplace=True)
```

```python
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
df['Phone_Number'].replace('nan--','',inplace=True)
```

```python
df[['Address','Street','ZipCode']] = df['Address'].str.split(',',n=2,expand=True)
```

```python
df['Address'].replace('Na',np.nan, inplace=True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a','')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
```

```python
df.drop('Not_Useful_Column',axis=1,inplace=True)
```

```python
df.replace('N/a','', inplace=True)
```

```python
df.fillna('',inplace=True)
```

```python
df
```

```python
for x in df['Do_Not_Contact'].values:
  if x == "Y":
    df = df[df['Do_Not_Contact'] != 'Y']
  elif x == '':
    df = df[df['Do_Not_Contact'] != '']
```

```python
for x in df.index:
  if df.loc[x,'Phone_Number'] == '':
    df.drop(x,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```
